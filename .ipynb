```python
import yfinance as yf
import datetime as dt
import pandas as pd
import numpy as np
```

```python
#list of stocks to download
stocks = ['GOOG','MSFT','MO','NVDA','AMD','TSLA','META']

#data for 360days back from todays
end_date =dt.datetime.today().strftime('%Y-%m-%d')


start_date = (dt.datetime.today()-dt.timedelta(360)).strftime('%Y-%m-%d')
print("Start Date:", start_date)
print("End Date:", end_date)
```

```python
# Generate trading days
trading_days = pd.bdate_range(start=start_date, end=end_date) #business date

print(f"Expected trading days: {len(trading_days)}")
```

```python
data=yf.download(stocks,start=start_date,end=end_date)
display(data)
```

```python
data.reset_index(inplace=True) #reset_index to make ticker a col and as well date col
print(data.index)
print(data.columns)
```

 'Price' and 'Ticker' are used as the two-level index for the column names
 => combine the two levels of column names into a single level.
 => Flatten the col names


```python
data.columns=['_'.join(col) for col in data.columns.values]
print(data.columns)
```

```python
# Melt the DataFrame to long format
data_melted = data.melt(id_vars=['Date_'],
                         var_name='Attribute_Ticker',
                                     value_name='Value')

# Split the 'Attribute_Ticker' column back into 'Attribute' and 'Ticker'
data_melted[['Attribute', 'Ticker']] = data_melted['Attribute_Ticker'].str.split('_', expand=True)

# Drop the 'Attribute_Ticker' column
data_melted.drop(columns=['Attribute_Ticker'], inplace=True)

data_melted.rename(columns={'Date_':'Date'}, inplace=True)
print(data_melted.head())
```

```python
df_pivot=data_melted.pivot_table(index=['Date', 'Ticker'], columns='Attribute', values='Value', aggfunc='first')
df_pivot.reset_index(inplace=True)
print(df_pivot.tail())
```

compute the 50-day and 200-day moving averages

```python
def Moving_average(ticker,df,ax):
    df=df.set_index('Date')
    ticker_df=df[df['Ticker']==ticker].copy()
    ticker_df['10MA']=ticker_df['Close'].rolling(window=10).mean()
    ticker_df['50MA']=ticker_df['Close'].rolling(window=50).mean()
    ticker_df['200MA']=ticker_df['Close'].rolling(window=200).mean()

    ax.plot(ticker_df.index, ticker_df['Close'], label='Close')
    ax.plot(ticker_df.index, ticker_df['10MA'], label='10-Day MA')
    ax.plot(ticker_df.index, ticker_df['50MA'], label='50-Day MA')
    ax.plot(ticker_df.index, ticker_df['200MA'], label='200-Day MA')
        
        # Set title and labels
    ax.set_title(f'{ticker} - Close and Moving Averages', fontsize='20')
    ax.set_xlabel('Date')   
    ax.set_ylabel('Price')
        
    # Add legend and grid
    ax.legend(fontsize='20')
    ax.grid(True)
    ax.tick_params(axis='x', rotation=45)  # Rotate x-axis labels for better readability
```

```python

def subplots(tickers,df,func):
    fig, axes=plt.subplots(3,int(len(tickers)/2), figsize=(50,8*len(tickers)))

    axes=axes.flatten()

    if len(tickers)==1:
        axes=[axes] #if only 1 ticker -> make it an array
    
    for i, ticker in enumerate(tickers):
        func(ticker, df, axes[i])
    
    # Adjust layout to avoid overlapping labels
    plt.tight_layout()
    plt.show()
```

```python
subplots(stocks, df_pivot,Moving_average)
```

### Calculate the expected returns and volatility

```python
# Calculate expected daily returns (mean of daily returns)
daily_returns = return_df.mean() #=> series

# Calculate annualized return for each stock
#numpy array of expected returns for the assets
annualized_return = (1 + daily_returns) ** len(trading_days) - 1

# Calculate annualized volatility for each stock
annualized_volatility = return_df.std() * np.sqrt(len(trading_days))

annualized_cov=return_df.cov()*(len(trading_days))

print(f'Annualized Return: {annualized_return}\n')
print(f'Annualized Volatility: {annualized_volatility}')
print(f'Annualized Cov: {annualized_cov}\n')
```

High Covariance Pairs:
- AMD and NVDA (0.149518): These two have relatively high covariance, consistent with their strong positive correlation. Both are in the semiconductor industry and likely influenced by similar market forces.
- TSLA (0.415508) and TSLA (itself): This diagonal element represents the variance of TSLA, showing its high standalone volatility

Low or Negative Covariance Pairs:
- MO and Most Stocks: MO has negative or near-zero covariance with other stocks, particularly AMD (-0.012139) and META (-0.010969). This confirms MOâ€™s potential for diversification in a portfolio dominated by tech stocks.
- GOOG and MO (-0.001240): A near-zero covariance indicates little relationship between these two stocks

Notable Pairings:
- AMD and TSLA (0.081930): While the covariance is positive, it is relatively low, suggesting these two assets might not strongly co-move despite being growth-focused stocks

High-Covariance Assets (e.g., AMD and NVDA): Including both in a portfolio may increase risk due to their tendency to move together.

Low or Negative-Covariance Assets (e.g., MO and tech stocks): Adding MO could reduce portfolio risk through diversification.

TSLAâ€™s High Variance: TSLA exhibits high standalone risk, which could significantly influence portfolio volatility if not balanced with lower-variance assets like MO.

### Generate a series of random portfolios to identify the efficient propotion

```python
def portfolio_performance(weights, returns,annualized_cov):
    #np.dot() dot product of two arrays
    #portfolio return = total sumation of weights*returns of each stock
    portfolio_return=weights@returns #1x4@4x1 #weight array for list of stocks and returns array as well
    portfolio_volatility=np.sqrt(weights.T@(annualized_cov@weights))
    return portfolio_return, portfolio_volatility

#create an empty 3x1000 matrix to store the result
result=np.zeros((3,1000))

#Monte Carlo Loop
for i in range(1000):
    #random array with as many stocks as there are 
    weights=np.random.random(len(stocks)) #numpy array of weights (0 -> 1) assigned to each asset in the portfolio 
    #Normalize weights so that their sum equals 1
    weights/=np.sum(weights)# Divide each weight by the sum of the weights normalizes them

    portfolio_return, portfolio_volatility=portfolio_performance(weights, annualized_return, annualized_cov)

    result[0,i]=portfolio_return
    result[1,i]=portfolio_volatility
    result[2,i]=portfolio_return/portfolio_volatility #SHARPE RATIO

#Find the index of the max Sharpe Ratio
max_sharpe_idx=np.argmax(result[2, :]) #=> identify the col of max sharpe
max_sharpe_return=result[0, max_sharpe_idx]
max_sharpe_volatility=result[1,max_sharpe_idx]
max_sharpe_ratio=result[2,max_sharpe_idx]

#Find the index of the min Volatility
min_volatility_idx=np.argmin(result[1, :]) #=> identify the col of min volatility
min_volatility=result[1, min_volatility_idx]
min_volatility_return=result[0, min_volatility_idx]

plt.figure(figsize=(10, 7))
plt.scatter(result[1,:], result[0,:], c=result[2,:], cmap='YlGnBu', marker='o')
plt.colorbar(label='Sharpe Ratio')

#Highliht the max Sharpe ratio and min volatility
plt.scatter(max_sharpe_volatility, max_sharpe_return, color='red', marker='*',  label='Max Sharpe Ratio')
plt.scatter(min_volatility, min_volatility_return, color='blue', marker='*', label='Min Volatility')

plt.title('Efficient Frontier')
plt.xlabel('Volatility (Standard Deviation)')
plt.ylabel('Expected Return')
plt.grid(True)
plt.show()

# Print details for both portfolios
print('Maximum Sharpe Ratio Portfolio:')
print(f'Volatility: {max_sharpe_volatility:.2f}')
print(f'Expected Return: {max_sharpe_return:.2f}')
print(f'Max Sharpe Ratio: {max_sharpe_ratio:.2f}')
print('\nMinimum Volatility Portfolio:')
print(f'Volatility: {min_volatility:.2f}')
print(f'Expected Return: {min_volatility_return:.2f}')
```

```python
max_sharpe_weights=np.zeros(len(stocks))
for i in range(1000):
    #random array with as many stocks as there are 
    weights=np.random.random(len(stocks)) #numpy array of weights (0 -> 1) assigned to each asset in the portfolio 
    #Normalize weights so that their sum equals 1
    weights/=np.sum(weights)# Divide each weight by the sum of the weights normalizes them

    portfolio_return, portfolio_volatility=portfolio_performance(weights, annualized_return, annualized_cov)

    if result[2,i] == max_sharpe_ratio:
        max_sharpe_weights=weights
        break

portfolio_weights_df=pd.DataFrame({'Ticker': stocks,
                                  'Weight': max_sharpe_weights})
portfolio_weights_df
```
